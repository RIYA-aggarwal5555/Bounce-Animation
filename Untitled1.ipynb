Gemini pro: chat bot w/image

loading the image

```python
import os
from PIL import Image

image_path = input('Enter image path:')

if not os.path.isfile(image_path):
  raise SystemExit("Invalid Image Path! Exiting....")


img = Image.open(image_path)
img
```

creating the image chatbot

```python
import time

model = gai.GenerativeModel('gemini-1.5-flash')

while True:
  prompt = input('User:')

  if prompt.lower() not in ['exit','quit','bye']:
    response = model.generate_content([prompt, img])

    print('')
    print(response.text)
    print('=' * 100)

  else:
    print('Exitting....')
    time.sleep(2)
    print('see ya....')
    break
```
